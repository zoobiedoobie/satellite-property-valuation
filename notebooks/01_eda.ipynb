```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load raw data
train_df = pd.read_csv("../data/train.csv")
test_df  = pd.read_csv("../data/test.csv")

train_df.shape
train_df.head()
train_df.info()
train_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(6,4))
sns.histplot(train_df["price"], bins=50, kde=True)
plt.title("Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(np.log1p(train_df["price"]), bins=50, kde=True)
plt.title("Log-Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=train_df["sqft_living"], y=train_df["price"], alpha=0.3)
plt.title("Sqft Living vs Price")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=train_df["grade"], y=train_df["price"])
plt.title("Grade vs Price")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(x=train_df["waterfront"], y=train_df["price"])
plt.title("Waterfront vs Price")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = train_df.corr(numeric_only=True)
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

## Exploratory Data Analysis (EDA) Observations

- The target variable (price) is heavily right-skewed, indicating the need for a log transformation before modeling.
- Square footage of living area (`sqft_living`) shows a strong positive correlation with property price.
- Higher construction quality (`grade`) leads to significantly higher property values.
- Waterfront properties have a clear price premium compared to non-waterfront properties.
- Location features (latitude and longitude) indicate spatial dependence in property pricing.
