```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("../data/train.csv")
test_df  = pd.read_csv("../data/test.csv")

print(train_df.shape, test_df.shape)
```

```python
TARGET = "price"

y = train_df[TARGET]
X = train_df.drop(columns=[TARGET])
```

```python
DROP_COLS = ["id", "date"]

X = X.drop(columns=DROP_COLS)
test_df = test_df.drop(columns=DROP_COLS)
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns

# Train
X[num_cols] = X[num_cols].fillna(X[num_cols].median())
X[cat_cols] = X[cat_cols].fillna("Unknown")

# Test
test_df[num_cols] = test_df[num_cols].fillna(X[num_cols].median())
test_df[cat_cols] = test_df[cat_cols].fillna("Unknown")
```

```python
from sklearn.preprocessing import LabelEncoder

for col in cat_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    test_df[col] = le.transform(test_df[col])
```

```python
train_processed = X.copy()
train_processed[TARGET] = y

train_processed.to_csv("../data/train_processed.csv", index=False)
test_df.to_csv("../data/test_processed.csv", index=False)

print("✅ Preprocessing complete. Files saved.")
```
